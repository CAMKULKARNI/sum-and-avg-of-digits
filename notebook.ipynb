```python
import matplotlib.pyplot as plt
```

```python
def length_of_number(n):
    return len(str(n))

def sum_of_digits(n):
    return sum([int(i) for i in str(n)])
```

```python
freq = [0 for _ in range(10)]
limit = 10000
for i in range(1, limit):
    s = sum_of_digits(i)
    l = length_of_number(s)
    while l > 1:
        s = sum_of_digits(s)
        l = length_of_number(s)
    freq[s] += 1

print(freq)
plt.plot(range(10), freq)
plt.show()       
```

```python
avg = []
limit = 10000
for i in range(1, limit):
    s = sum_of_digits(i)
    l = length_of_number(s)
    avg.append(s / l)

plt.hist(avg)
plt.show()       
```
